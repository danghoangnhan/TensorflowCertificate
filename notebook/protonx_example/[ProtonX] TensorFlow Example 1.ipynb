### Phần lập trình 

````python
# Import thư viện cần thiết 
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
````

````python
# Dữ liệu 
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-7.0, -3.0, 1.0, 5.0, 9.0, 13.0], dtype=float)

max_features, min_features = max(xs), min(xs)
features = (xs-min_features)/(max_features-min_features)
````

````python
# Định nghĩa Model cho bài toán Linear Regression 
model = Sequential()
model.add(Dense(units=1, input_shape=[1]))     
````

````python
# Train model 
model.compile("adam", loss='mean_squared_error', metrics=['mean_squared_error'])
model.fit(xs, ys, epochs=1000, verbose=1)
````

````python
# Lưu Model và nộp 
model.save("mymodel.h5")
````

````python
print(model.predict([-1]))
````
