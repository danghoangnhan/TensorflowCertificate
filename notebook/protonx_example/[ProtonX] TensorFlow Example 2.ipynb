````python
# TensorFlow and tf.keras
import tensorflow as tf

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import SGD
print(tf.__version__)
````

````python
fashion_mnist = tf.keras.datasets.mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
````

````python
# Write your code here 
train_images = train_images / 255
test_images = test_images / 255
train_images.shape
````

````python
model =  tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))  
model.add(tf.keras.Dense(32, activation='relu'))  
model.add(tf.keras.Dense(64, activation='relu'))  
model.add(tf.keras.Dense(32, activation='relu')) 
model.add(tf.keras.Dense(10, activation='softmax'))
````

````python
# Train model 
model.compile(SGD(0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=1000, batch_size=256, verbose=1, validation_data=(test_images, test_labels))
````

````python
# Lưu Model và nộp 
model.save("mymodel.h5")
````
